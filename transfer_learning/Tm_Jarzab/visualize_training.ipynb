```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
import os
import pickle
```

```python
def plot_training(infile,set_xlim=False,set_ylim=True,test=False):
    df = pd.read_csv(infile,index_col=0)
    loss = df['val_loss'].values
    r2 = df['val_coef_det_k'].values
    try:r2_test = df['test_val_det_k'].values
    except: pass
    print('At epoch {0}/{1}'.format(df.index[np.argmin(loss)],df.index[-1]))
    print('Best Val R2  : {0:.2f}'.format(r2[np.argmin(loss)]))
    print('Best Val RMSE: {0:.2f}'.format(np.min(loss)**0.5))
    try:
        print('Test R2      : {0:.2f}'.format(r2_test[np.argmin(loss)]))
        print('Test RMSE    : {0:.2f}'.format(df['test_loss'][np.argmin(loss)]**0.5))
    except: pass
    
    plt.figure(figsize=(7,3))
    plt.subplot(1,2,1)
    plt.scatter(df.index,df['coef_det_k'],label='Training')
    plt.scatter(df.index,df['val_coef_det_k'],label='Validation')
    if test: plt.scatter(df.index,df['test_val_det_k'],label='Test')
    #except: pass
    if set_ylim: plt.ylim(0,1)
    #plt.legend()
    #if patience == 200: plt.xlim((0,df.index[-1]-150))
    plt.xlabel('Epochs')
    plt.ylabel('$R^2$')
    plt.legend()
    if set_xlim:plt.xlim(0,400)
    #plt.title('Size: {0}k'.format(s))
    
    plt.subplot(1,2,2)
    plt.scatter(df.index,np.sqrt(df['loss']),label='Training')
    plt.scatter(df.index,np.sqrt(df['val_loss']),label='Validation')
    if test: plt.scatter(df.index,np.sqrt(df['test_loss']),label='Test')
    #except: pass
    if set_ylim: plt.ylim([0,20])
    #if patience == 200: plt.xlim((0,df.index[-1]-150))
    plt.legend()
    #plt.title('Size: {0}k'.format(s))
    
    # print the performance of the model with the highest validation score
    sort_ind = np.argsort(df['val_coef_det_k'].values)
    plt.xlabel('Epochs')
    plt.ylabel('RMSE ($^\circ$C)')
    if set_xlim:plt.xlim(0,400)
    plt.tight_layout()
    plt.show()
    
    try: return r2[np.argmin(loss)], r2_test[np.argmin(loss)]
    except: return r2[np.argmin(loss)]
    
    
```

```python
indir = '../../results/tm_Jarzab_models/TL6_RES1_ogt_onehot_SameAnces'
for name in os.listdir(indir):
    if not name.endswith('.csv'): continue
    #if 'fix' not in name: continue
    #if '1e' not in name: continue
    print(name)
    try:plot_training(os.path.join(indir,name))
    except: None
    print()
```
